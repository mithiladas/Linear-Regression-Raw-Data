Efficient Linear Regression using Numpy 

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([20000, 30000, 40000, 50000])
y = np.array([40000, 60000, 80000, 100000])

#x = np.array([1,2,3,4])
#y = np.array([2,3,5,4])


n = len(x)

mean_x = np.mean(x)
mean_y = np.mean(y)

numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x) ** 2)
m = numerator / denominator
b = mean_y - m * mean_x

print("Linear Regression: y = {:.2f} * x + {:.2f}".format(m, b))

y_pred = m * x + b
errors = (y - y_pred) ** 2

print("\nError at each stage")
for i in range(n):
    print(f"x = {x[i]}, y = {y[i]}, ŷ = {y_pred[i]:.2f}, Error = {y[i] - y_pred[i]:.2f}, Squared Error = {errors[i]:.2f}")

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.scatter(x, y, color='blue', label='Actual data')
plt.plot(x, y_pred, color='red', label='Fitted line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Linear Regression Fit')

plt.subplot(1, 2, 2)
plt.plot(range(1, n+1), errors, marker='o', linestyle='--', color='green')
plt.xlabel('Data Point Index')
plt.ylabel('Squared Error')
plt.title('Squared Error per Data Point')
plt.grid(True)

plt.tight_layout()
plt.show()
```

Linear Regression Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt

#x = np.array([20000, 30000, 40000, 50000])
#y = np.array([40000, 60000, 80000, 100000])

x = np.array([1,2,3,4])
y = np.array([2,3,5,4])


x_normalized = (x - np.mean(x)) / np.std(x)
y_normalized = (y - np.mean(y)) / np.std(y)


m = 0.0
b = 0.0
learning_rate = 0.1
epochs = 100

n = len(x)
history_m = []
history_b = []
history_cost = []

for epoch in range(epochs):
    y_pred = m * x_normalized + b
    error = y_pred - y_normalized
    
    cost = np.sum(error ** 2) / n
    history_cost.append(cost)
    
    dm = (2/n) * np.sum(error * x_normalized)
    db = (2/n) * np.sum(error)
    
    m -= learning_rate * dm
    b -= learning_rate * db
    
    history_m.append(m)
    history_b.append(b)

m_original = m * (np.std(y) / np.std(x))
b_original = np.mean(y) - m_original * np.mean(x)

print(f"Gradient Descent Result: y = {m_original:.2f} * x + {b_original:.2f}")

y_pred_original = m_original * x + b_original
errors = (y - y_pred_original) ** 2

print("\nError at each stage")
for i in range(n):
    print(f"x = {x[i]}, y = {y[i]}, ŷ = {y_pred_original[i]:.2f}, Squared Error = {errors[i]:.2f}")

fig, axes = plt.subplots(2, 2, figsize=(10, 8))


axes[0, 0].scatter(x, y, color='blue')
axes[0, 0].plot(x, y_pred_original, color='red')
axes[0, 0].set_xlabel('X')
axes[0, 0].set_ylabel('Y')
axes[0, 0].set_title('Linear Regression Fit')

axes[0, 1].plot(history_cost, color='green')
axes[0, 1].set_xlabel('Epoch')
axes[0, 1].set_ylabel('Cost (MSE)')
axes[0, 1].set_title('Cost Function over Epochs')
axes[0, 1].grid(True)

axes[1, 0].plot(history_m, label='m (slope)', color='orange')
axes[1, 0].set_xlabel('Epoch')
axes[1, 0].set_ylabel('m value')
axes[1, 0].set_title('Slope (m) Convergence')
axes[1, 0].legend()
axes[1, 0].grid(True)

axes[1, 1].plot(history_b, label='b (intercept)', color='purple')
axes[1, 1].set_xlabel('Epoch')
axes[1, 1].set_ylabel('b value')
axes[1, 1].set_title('Intercept (b) Convergence')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

OLS with Gradient Descent

```python
import numpy as np
import matplotlib.pyplot as plt


X = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])


m = 0          
b = 0          
learning_rate = 0.01       
epochs = 1000  
n = float(len(X)) 


losses = []

for i in range(epochs): 
    y_pred = m * X + b              
    error = y - y_pred              
    mse = np.mean(error**2)         
    losses.append(mse)
    
    
    dm = (-2/n) * sum(X * error) 
    dc = (-2/n) * sum(error)
    
    
    m = m - learning_rate * dm
    b = b - learning_rate * dc

print(f"Final Slope (m): {m:.2f}")
print(f"Final Intercept (b): {b:.2f}")


plt.scatter(X, y, color='red', label='Actual Data') 
plt.plot(X, m*X + b, color='blue', label='OLS Best Fit Line')
plt.xlabel('X (Independent Variable)')
plt.ylabel('y (Dependent Variable)')
plt.legend()
plt.title('OLS Regression - Error Minimization')
plt.show()
```
